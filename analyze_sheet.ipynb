```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_excel('detritus.xlsx')
```

```python
pd.options.plotting.backend = "plotly"
```

```python
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
df = df.iloc[: , 3:]
```

```python
df
```

```python
color=[df['Ground Truth'][i + 1] + str(df['Wrong classifications'][i + 1]) for i in range(len(df['Ground Truth']))]
```

```python
fig = px.scatter(x=df['mobilenet_x'], y=df['mobilenet_y'], color=color)
fig.show()
```

```python
fig = px.scatter(x=df['densenet_x'], y=df['densenet_y'], color=color)
fig.show()
```
